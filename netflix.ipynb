```python
#content distributions
df['type'].value_counts()
```

```python
sns.countplot(x="type",data=df)
plt.show()
```

```python
#percentage
percentage = (df['type'].value_counts(normalize=True)*100).round(2)
percentage
```

```python
df['type'].value_counts().plot(kind="pie",autopct="%1.1f%%")
plt.title("Percentage of Movies vs TV Shows")
plt.ylabel("")
plt.show()
```

```python
#Trend Over Time
year_counts = df['year_added'].value_counts().sort_index()
year_counts
```

```python
plt.figure()
year_counts.plot(kind="line")
plt.title("Netflix Content Added over years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
df["country"].value_counts().sort_values()
```

```python
df['country'].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Content Producing Countries")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

```python
#Rating Counts
rating_counts = df["rating"].value_counts()
rating_counts
```

```python
plt.figure()
rating_counts.plot(kind="bar")
plt.title("Distributions of Netflix content Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Title")
plt.xticks(rotation = 45)
plt.show()
```

```python
#Genre Analysis
genre = df["listed_in"].value_counts().head(10)
genre
```

```python
from collections import Counter
```

```python
all_genres = df['listed_in'].str.split(",").sum()
genre_counts = Counter(all_genres)
genre_counts.most_common(10)
```

```python
top_genres = pd.Series(genre_counts).sort_values(ascending=False).head(10)

plt.figure()
top_genres.plot(kind="bar")
plt.title("Top 10 Genre on Netflix")
plt.xlabel("Genre")
plt.ylabel("Number of Titles")
plt.xticks(rotation=75)
plt.show()
```

# Key Insights

A. Content Distributions 
  - Netflix catalog is dominated by movies,showing a strategy to maintain large and diverse content volume.
  - TV shows are fewer but strategically important because they increase watch time and long-term user retention.
  - Around 70% movies and 30% TV shows indicates a movie strategy.
  - Growing TV show share reflects shift towards binge-watch culture.

B. Trend Over Time
  - Rapid growth after 2015 shows Netflix's global expansion and heavy investment in originals.
  - Peak additions near 2018-2020 indicate aggresive market capture before saturation
  - Slight stabilization later suggests mature content stategy.

C. Country-wise Analysis
  - USA leads production,confirming it as Netflix's primary content hub.
  - Strong presence of India, UK and others highlights localization strategy for global subscibers.

D. Rating Analysis
  - Dominance of TV-MA and TV-14 shows focus on teen and adult audiences,the highest engagement segment.
  - Limited kids only content suggests selective diversification rather than core targeting.

E. Genre Analysis
  - Drama and Comedy lead due to
    
    1. High emotional connection
    2. Global relatability
    3. Strong repeat-watch value

## Conclusion

This EDA reveals how Netflix uses data-driven content distribution,regional production,audience targeting,and genre strategy to maximize engagement,retention and global expansion
